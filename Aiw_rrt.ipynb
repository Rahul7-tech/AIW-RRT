```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Simulate exploration process (simplified)
def compute_information_gain(tree, environment):
    # Simulate entropy calculation and information gain
    entropy = np.random.uniform(0, 1, size=len(environment))
    return np.sum(entropy)

def aiw_rrt_algorithm(environment, num_steps=100):
    # Adaptive Information Gain Weighted Rapidly Exploring Random Tree (AIW-RRT)
    info_gain = []
    for step in range(num_steps):
        # Simulate tree expansion and information gain calculation
        new_info_gain = compute_information_gain([], environment)
        info_gain.append(new_info_gain)
    return info_gain

def rrt_algorithm(environment, num_steps=100):
    # Rapidly Exploring Random Tree (RRT)
    info_gain = []
    for step in range(num_steps):
        # Simulate tree expansion and information gain calculation
        new_info_gain = np.random.uniform(0, 0.5)  # Simpler random gain for RRT
        info_gain.append(new_info_gain)
    return info_gain

def rrt_star_algorithm(environment, num_steps=100):
    # Rapidly Exploring Random Tree Star (RRT*)
    info_gain = []
    for step in range(num_steps):
        # Simulate tree expansion and information gain calculation with RRT*
        new_info_gain = np.random.uniform(0.2, 0.7)  # Different range for RRT*
        info_gain.append(new_info_gain)
    return info_gain

# Simulate a random environment (for demonstration purposes)
environment = np.random.rand(100, 2)  # 100 random points in 2D space

# Run all three algorithms
num_steps = 100
aiw_rrt_info_gain = aiw_rrt_algorithm(environment, num_steps)
rrt_info_gain = rrt_algorithm(environment, num_steps)
rrt_star_info_gain = rrt_star_algorithm(environment, num_steps)

# Plotting histogram comparison of information gain for the three algorithms
plt.figure(figsize=(10, 6))

plt.hist(aiw_rrt_info_gain, bins=20, alpha=0.5, label="AIW-RRT", color='b')
plt.hist(rrt_info_gain, bins=20, alpha=0.5, label="RRT", color='r')
plt.hist(rrt_star_info_gain, bins=20, alpha=0.5, label="RRT*", color='g')

plt.title('Histogram Comparison of Information Gain for AIW-RRT, RRT, and RRT*')
plt.xlabel('Information Gain')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
